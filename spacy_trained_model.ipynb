```python
file = "Bank_Statements/axis1bs.pdf"
```

```python
import pdfplumber
def totxt(file):
    with pdfplumber.open(file) as pdf:

        print(pdf.pages)
        print(type(pdf.pages))
        txt = ""
        for i in range(len(pdf.pages)):
            first_page = pdf.pages[i]
            txt += first_page.extract_text()
        txt = txt.replace("(cid:9)", "")
        return txt
```

```python
nlp = spacy.load("model_best")
```

```python
doc = nlp(totxt(file))
```

```python
from spacy import displacy
colors = {'NAME': "#85C1E9", "ACCOUNT_TYPE": "#ff6961","BASE_BRANCH" :"#FFFF00","BANK":"#00FFFF","PARTICULARS":"#FFEBCD"
         ,"IFSC":"#D2691E",'HOLDING TYPE':"#DAA520",'ACCOUNT NO':"#00FF7F"}
options = {"ents": ['NAME', 'ACCOUNT_TYPE',"BASE_BRANCH",'BANK','PARTICULARS','IFSC','HOLDING TYPE','ACCOUNT NO'], "colors": colors}

displacy.render(doc, style='ent',options=options)
```
